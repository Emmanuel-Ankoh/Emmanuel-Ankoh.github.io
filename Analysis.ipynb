## Data preparation

### Task:
Clean the 'aqbce' column in all three dataframes to prepare for merging.  Handle missing values in other key columns as needed.


**Reasoning**:
Clean the 'aqbce' column in all three dataframes and handle missing values in the 'beneficiario' column of df_payment.



```python
# Clean 'aqbce' column
for df in [df_practise, df_payment, df_companies]:
    df['aqbce'] = df['aqbce'].astype(str).str.strip().str.upper()

# Handle missing values in 'beneficiario'
# Impute missing 'beneficiario' values with 'Unknown'
# Rationale:  Imputation with a placeholder value is chosen as it preserves the rows with other potentially useful data.
df_payment['beneficiario'] = df_payment['beneficiario'].fillna('Unknown')
```

## Data wrangling

### Task:
Merge the three dataframes: `df_practise`, `df_payment`, and `df_companies` based on the cleaned 'aqbce' column.


**Reasoning**:
Merge the three dataframes based on the 'aqbce' column using inner joins.



```python
# Merge df_practise and df_payment
merged_df1 = pd.merge(df_practise, df_payment, on='aqbce', how='inner')

# Merge merged_df1 and df_companies
merged_df = pd.merge(merged_df1, df_companies, on='aqbce', how='inner')

# Inspect the shape of the final merged DataFrame
print("Shape of merged_df:", merged_df.shape)
display(merged_df.head())
```
